### Build the Model

```python
X = data.drop("target", axis = 1)
y = data["target"]
```

### Spliting Dataset into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=10)
```

### Model Building

#### a. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
pred_logistic=log_reg.predict(X_test)
```

```python
print(classification_report(y_test, pred_logistic))
print(confusion_matrix(y_test, pred_logistic))

logistic_score =accuracy_score(pred_logistic,y_test)*100 
print('Logistic Regression Score: ',logistic_score)
```

```python
cm = confusion_matrix(y_test,pred_logistic)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="Greens");
```

#### b. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtr = DecisionTreeClassifier()
```

```python
dtr.fit(X_train,y_train)
```

```python
pred_dtr = dtr.predict(X_test)
```

```python
print(classification_report(y_test, pred_dtr))
print(confusion_matrix(y_test, pred_dtr))

dtr_score = accuracy_score(pred_dtr,y_test)*100
print('Decision Tree Score: ',dtr_score)
```

```python
cm = confusion_matrix(y_test,pred_dtr)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="Greens");
```

#### c. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=10,max_depth=10,n_estimators=50)
```

```python
rf.fit(X_train,y_train)
```

```python
pred_rf = rf.predict(X_test)
```

```python
print(classification_report(y_test, pred_rf))
print(confusion_matrix(y_test, pred_rf))

rf_socre = accuracy_score(pred_rf,y_test)*100
print('Random Forest Score: ',rf_socre)
```

```python
cm = confusion_matrix(y_test,pred_rf)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="Greens");
```

#### d. Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(X_train, y_train)
```

```python
pred_svm = svm.predict(X_test)
```

```python
print(classification_report(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))

svm_socre = accuracy_score(pred_svm,y_test)*100
print('Support Vector Machine Score: ',svm_socre)
```

```python
cm = confusion_matrix(y_test,pred_svm)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="Greens");
```

#### e. K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 ) 
```

```python
knn.fit(X_train, y_train)
```

```python
pred_knn = knn.predict(X_test)
```

```python
print(classification_report(y_test, pred_knn))
print(confusion_matrix(y_test, pred_knn))

knn_list=[]
for i in range(2,21):
    knn2=KNeighborsClassifier(n_neighbors=i)
    knn2.fit(X_train,y_train)
    knn_list.append(knn2.score(X_test,y_test))
    
plt.plot(range(2,21),knn_list) 
plt.xlabel("K Value")
plt.ylabel("Score")
plt.show()

knn_socre = max(knn_list)
print('K-Nearest Neighbour Score: ',knn_socre*100)
```

### Model Comparision

```python
compare = pd.DataFrame({'Model': ['Random Forest', 'Logistic Regression', 'Decision Tree', 'Support Vector Machine', 'K Neighbors'], 
                        'Accuracy': [ rf_socre, logistic_score, dtr_score, svm_socre, knn_socre*100]})
compare.sort_values(by='Accuracy', ascending=False)
```
